```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
```

```python
x
```

```python
y
```

```python
# Graph controlling functions:- 
colours = np.random.rand(50)
sizes = 100*np.random.rand(50)
plt.scatter(x,y, c = colours, s = sizes, alpha = .5) # 'Alpha' represents the intensity of colours and range(0-1)
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.title("Random data")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = '#FF5733', s = sizes, alpha = .5,marker='v')  # 'Alpha' represents the intensity of colours and range(0-1)
plt.xlabel("x-Axis")  
plt.ylabel("y-Axis")
plt.title("Random data")
# marker is used to show the data in different shapes.
```
